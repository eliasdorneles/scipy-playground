### Testing different colormaps

```python
num_elems=50
figsize=(8, 4)

labels = np.arange(num_elems)
colormaps = [plt.cm.Spectral, plt.cm.rainbow, plt.cm.Set1, plt.cm.Accent, plt.cm.Accent_r]

fig, ax = plt.subplots(figsize=figsize)
for y, cmap in enumerate(colormaps, 1):
    colors = cmap(np.linspace(0, 1, len(labels)))
    for x, (color, label) in enumerate(zip(colors, labels), 1):
        plt.plot(x, y, 'o', markerfacecolor=color, markersize=6)
        ax.text(len(labels) + 1, y, cmap.name, size=12, ha='left', va='bottom')
plt.ylim(0, len(colormaps) + 1)
plt.xlim(0, len(labels) + 8)
```

### Creating a generic complex plot function

```python
def complex_plot(x, y, size=None, color=None):
    x, y = np.asarray(x), np.asarray(y)
    if size is not None:
        #TODO: normalize between 3 and 20 (?)
        size = np.asarray(size)
    else:
        size = np.ones_like(x) + 5
    if color is not None:
        #TODO: normalize between 0 and 1
        color = np.asarray(color, dtype='f')
        color = (color - color.min()) / (color.max() - color.min())
        print color
    else:
        color = np.ones_like(x)
    for i, j, k, c in zip(x, y, size, plt.cm.rainbow(color)):
        plt.plot(i, j, 'o', markersize=k, markerfacecolor=c)
```
